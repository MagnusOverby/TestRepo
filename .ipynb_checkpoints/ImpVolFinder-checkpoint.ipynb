```python
class ImpliedVol:
    def __init__(self,K,T,vol,openInt):
        self.T=int(T)
        self.K=float(K)
        self.vol=vol
        self.openInt = openInt
    def T(self):
        return self.T
    def K(self):
        return self.K
    def vol(self):
        return self.vol
    def openInt(self):
        return self.openInt
```

```python
def GetMats(ticker,soup):
    tenors = []
    for tag in soup.findAll("option"):
        tmp = float(tag.attrs.values()[1])
        tenors.append(tmp)
    return sorted(np.array(tenors),reverse=True)
def getStockPriceFromDump(ticker,soup):
  for x in soup.find_all("span"):
    tmp = x.attrs
    val = x.text
    if tmp.has_key('data-sq'):  
      if tmp['data-sq'] == ticker+':value':
        return float(val)
def getInitialSoup(ticker):
    path = "http://finance.yahoo.com/q/op?s=[CHANGE]&date="
    path = path.replace('[CHANGE]',ticker)
    htmltext= urllib.urlopen(path).read()
    soup = BeautifulSoup(htmltext)
    return soup
```

```python
# takes a list of ImpVol Points and transforms it into a dataframe
# takes wK -> wanted Strikes.. maybe a max / min strike is better
# takes wT -> wanted Tenors.. maybe a max / min tenor is better 
# takes S -> StockPrice
# takes minVolume -> minimal volume to be included 
# can be customized in a few ways
class rVolSurf:
    def __init__(self,Points,minStrike,maxStrike,minTen,S,minVolume):
        #long constructor hallmarks of a charlatan programmer !
        self.S = S
        
        self.minVolume = minVolume
        self.DefineStrikesNtenors(Points,minTen,minStrike,maxStrike)
        
        tmpVol = np.zeros([len(self.K),len(self.T)])
        for x in Points:
            if x.K in self.K:
                i = self.K.index(x.K)
                if x.T in self.T:
                    j = self.T.index(x.T)
                    tmpVol[i,j] = x.vol
        #Vols
        self.vol = tmpVol
        self.SetDisplayStrikes()
        
    def DefineStrikesNtenors(self,Points,minTen,minStrike,maxStrike):
        wK=[]
        wT=[]
        #build the 3 different grids maybe move this out to an external function 
        for x in Points:
            #print x.K > minStrike
            #print "Strike: "+str(x.K)
            #print "minStrike: "+str(minStrike)
            #print "maxStrike: "+str(maxStrike)
            #print "IsIn? " + str((x.K > minStrike) and (x.K < maxStrike))
            if (x.K > minStrike and x.K < maxStrike):
                wK.append(x.K)
            if x.T > minTen:
                wT.append(x.T)
        wT = list(set(wT))
        wK = list(set(wK))
        wK.sort()
        wT.sort()
        self.K=wK
        self.T=wT
        
    def Conditions(self,point):
        return (self.minVolume > point.openInt)
    def SetDisplayStrikes(self):
        tmpK = [str(np.round((x/self.S)*100,1))+'%' for x in self.K]
        self.dK = tmpK
    def GetDK(self):
        return self.dK
    def GetDf(self):
        df = pd.DataFrame(self.vol,index=[self.dK],columns=[self.T])
        return df
    def GetK(self):
        return self.K
    def GetT(self):
        return self.T
```

```python
def GetVolList(ticker,unixtime):
    Basepath =  "http://finance.yahoo.com/q/op?s=[CHANGE]&date=[UNIXTIME]"
    tmp = datetime.datetime.fromtimestamp(float(unixtime))
    delta = tmp-datetime.datetime.today()
    D =  str(delta.days)
    #Would be nice to have spot as well to build a proper strike at once
    path =Basepath.replace('[CHANGE]',ticker)
    path =path.replace('[UNIXTIME]',unixtime)
    r = requests.get(path)
    soup2 = BeautifulSoup(r.content)
    data = np.array([x.text.encode('utf-8') for x in soup2.find_all("td")])
    data = data[2:]
    dMax = len(data)
    rowChange = False
    puts = []
    calls = []
    for i in range(1,len(data)):
        if '[modify]' in data[i]:
            rowChange = True
        if rowChange:
            puts.append(str(data[i]).replace('\n',''))
        else:
            calls.append(str(data[i]).replace('\n',''))
    Points =[]
    #add all calls
    callStrikes = calls[0::10]
    putStrikes = puts[1::10]
    callVol = np.array([float(x.replace('%',''))/100 for x in calls[9::10]])
    putVol =  np.array([float(x.replace('%',''))/100 for x in puts[10::10]])
    callVolume = calls[8::10]
    putVolume = puts[9::10]
    #add puts
    for i in xrange(len(callVol)):
        Points.append(ImpliedVol(callStrikes[i],D,callVol[i],callVolume[i]))
    #add calls
    for j in xrange(len(putVol)):
         Points.append(ImpliedVol(putStrikes[j],D,putVol[j],putVolume[j]))
    return Points
```

```python
#Returns a dataframe with Index Strikes (absolute) and Header Tenor
def GetMidVols(ticker,unixtime):
    Basepath =  "http://finance.yahoo.com/q/op?s=[CHANGE]&date=[UNIXTIME]"
    tmp = datetime.datetime.fromtimestamp(float(unixtime))
    delta = tmp-datetime.datetime.today()
    D =  str(delta.days)
    #Would be nice to have spot as well to build a proper strike at once
    path =Basepath.replace('[CHANGE]',ticker)
    path =path.replace('[UNIXTIME]',unixtime)
    r = requests.get(path)
    soup2 = BeautifulSoup(r.content)
    data = np.array([x.text.encode('utf-8') for x in soup2.find_all("td")])
    data = data[2:]
    dMax = len(data)
    rowChange = False
    puts = []
    calls = []
    for i in range(1,len(data)):
        if '[modify]' in data[i]:
            rowChange = True
        if rowChange:
            puts.append(str(data[i]).replace('\n',''))
        else:
            calls.append(str(data[i]).replace('\n',''))
    return DataUsage(puts,calls,D)
```

```python
#Fetch the HTML in this loop (will be alot more efficient !!)
#There are alot of int float conversions going on since we need int for the URL but float to convert to date.
def AddNeededTenors(neededTenors,Vals,headers,idxx):
    neededTenors = ['30','90','180','360']
    lMax=len(headers)
    headers.extend(neededTenors)
    headers = np.unique(headers)
    lInd = len(idxx)
    Carrier = pd.DataFrame(np.zeros([lInd,len(headers)]),columns=headers,index=idxx)
    for i in range(0,lMax):
        tmplen = len(Vals[i])
        tmp = Vals[i]
        for j in range(0,tmplen-1):
            k = idxx.index(tmp[j][1])
            Carrier.ix[k,i] = tmp[j][0]     
    tmpCar1 = Carrier
    tmpCar1.columns =  tmpCar1.columns.astype(float)
    tmpCar2= tmpCar1.T
    return tmpCar2.sort_index().T

def GetVolSurf(ticker,soup,neededTenors):
    Mat = GetMats(ticker,soup)
    minVolume = 500
    lMax = len(Mat)
    idx = []
    Vals = [[] for x in range(0,lMax)]
    headers = []
    allIndex = []#This will remove alot of interesting points... probably not ideal....
    #Idea:
    #1 build a set of all possible strikes
    #2 Interpolate (using some scheme) between awailable strikes.
    for i in range(0,lMax):
        tmp = str(int(Mat[i])) 
        tmpdf = GetMidVols(ticker,tmp)
        htmp = str(tmpdf.columns[0])
        headers.append(htmp)
        jMax = len(tmpdf.index)
        for j in range(0,jMax-1):
            idx.append(float(tmpdf.index[j]))
            if tmpdf.ix[j,0] != 0.0:
                Vals[i].append((tmpdf.ix[j,0],float(tmpdf.index[j])))              
    idxx = [x for x in set(idx)]
    idxx.sort()
    return AddNeededTenors(neededTenors,Vals,headers,idxx)
```

```python
#this is just a flat interpolation of the surface
def GetAdhocSurface(ticker,neddedTenors):
    soup = getInitialSoup(ticker)
    vArr = GetVolSurf(ticker,soup)
    vArr[vArr==0] = np.nan
    spot = getStockPriceFromDump(ticker,soup)
    newIndex = np.round((vArr.index / spot)*100,0)
    newIndex = [str(int(x))+'%' for x in newIndex]
    Interpolated =  vArr.interpolate().T.interpolate().T.ffill().bfill()
    Interpolated.index = newIndex
    return Interpolated

def GetAdhocSurfaceTest(ticker,neddedTenors):
    soup = getInitialSoup(ticker)
    vArr = GetVolSurf(ticker,soup,neddedTenors)
    vArr[vArr==0] = np.nan
    spot = getStockPriceFromDump(ticker,soup)
    newIndex = np.round((vArr.index / spot)*100,0)
    newIndex = [str(int(x))+'%' for x in newIndex]
    return vArr.T
```

```python
#the full class
class YahooVolSurface:
    def __init__(self,ticker,printTenors,wPath):
        self.ticker = ticker
        self.pTenors=printTenors
        self.rVolSurf=self.PullSurf()
        self.wPath = wPath
        
    def PullSurf(self):
        soup = getInitialSoup(self.ticker)
        vArr = GetVolSurf(self.ticker,soup,self.pTenors)
        vArr[vArr==0] = np.nan
        spot = getStockPriceFromDump(self.ticker,soup)
        newIndex = np.round((vArr.index / spot)*100,0)
        newIndex = [str(int(x))+'%' for x in newIndex]
        vArr.index = newIndex
        OutPut = vArr.T
        return OutPut
    def interpolation(self,method):
        if method == 'linear':
            tmp = self.rVolSurf
            self.iVolSurf= tmp.interpolate().T.interpolate().T.ffill().bfill()
            self.iVolSurf.T
                  
    def reduceStrikes(self,strikes):
        self.iVolSurf = self.iVolSurf[strikes]
        
    def GetSurf(self):
        return dropDuplicateStrikes(self)
    
    def dropDuplicateStrikes(self):
        tmpSurf = self.iVolSurf.T.copy()
        akk = tmpSurf.columns
        end = len(akk)
        idx = tmpSurf.index
        tmpSurf[9999.0] = idx
        tmpSurf.drop_duplicates(cols = [9999.0], inplace = True)
        self.iVolSurf = tmpSurf.drop(9999.0,1)
        return self.iVolSurf
    
    
    def WriteToFile(self):
        FileNmn = "GarantumVolStock.xls"
        #filess.to_csv(path+FileNmn,sep='\t',header=True,index=True,na_rep='')
        firstRow = ['Index Implied Vols %','','','','','','']
        Rows = [] 
        Rows.append(firstRow)
        for surface in VolFile:
            Rows.append([surface[0][0],surface[0][0]])
            Rows.append(surface[0])
            for row in surface[1:]:
                tmpR = [str(np.round(x,2)).replace('.',',') for x in row[2:]]
                tmp = [row[0],row[1]]
                tmp.extend(tmpR)
                Rows.append(tmp)
        df = pd.DataFrame(Rows)
        df.to_csv(self.wPath+FileNmn,sep='\t',header=False,index=False,na_rep='')
```
